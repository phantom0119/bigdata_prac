```python
import pandas as pd

file = pd.read_csv(r'G:\ETC\5.git_storage\bigdata_prac_clone\Regression\dataset\multiple_linear_regression_dataset.csv')
mdf = pd.DataFrame(file)

mdf.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# 그래프를 노트북 셀 안에 바로 표시
%matplotlib inline
```

```python
sns.lmplot(x='age', y='experience',
           data=mdf[['age', 'experience']],
           fit_reg=False,
           scatter_kws={'s': 25},
          height=3,
          aspect=1)
plt.show()
```

```python
# 데이터셋의 각 Column간 관계를 파악하기 위해 그래프를 그린다.
# 그래프 크기 설정: (가로, 세로)
fig, axes = plt.subplots(1, 3, figsize=(12, 4))  # 1행 3열, 서브플롯 크기 설정

sns.scatterplot(x='age', y='experience',
                data=mdf[['age', 'experience']],
                ax=axes[0])

sns.scatterplot(x='age', y='income',
                data=mdf[['age', 'income']],
                ax=axes[1])

sns.scatterplot(x='experience', y='income',
                data=mdf[['experience', 'income']],
                ax=axes[2])


plt.tight_layout()  # 레이아웃 자동 정렬
plt.show()
```

```python
# 독립변수 2개 (age, experience)
# 종속변수 1개 (income)
# 다중 선형 회귀 모델 연습

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

X = mdf[['age', 'experience']]
Y = mdf['income']

# 스케일러를 통해 평균 0, 표준편차 1의 특성(Feature) 변환을 수행.
# 스케일이 서로 다른 변수 간의 (ex. 나이(age)와 경력(experience)) 수치 차이를 제거해 비교 가능.
'''
경사하강법 기반 선형 회귀에서 빠르고 안정적인 학습 (수렴 속도 향상).
정규화 회귀(Lasso, Ridge), 가중치 패널티를 부여하는 회귀에서 특성의 스케일이 결과에 크게 영향을 준다.
각 변수의 상대적 중요도 비교에 용이.
'''
scaler = StandardScaler()
model = LinearRegression()


# 독립변수 스케일링
x_scaled = scaler.fit_transform(X)

# 예측(종속) 변수는 실제 예측하고자 하는 수치를 유지하기 위해 스케일링 X
x_train, x_test, y_train, y_test = train_test_split( x_scaled, Y,
                                                    random_state=55,
                                                    test_size=0.2)

print(x_train.shape, x_test.shape)


```

```python
# 모델 학습 진행
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
print(pred)
```

```python
print(y_test)
```

```python
# 모델 평가
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# MSE : 평균 제곱 오차
print(mean_squared_error(pred, y_test))

# MAE : 평균 절대 오차
print(mean_absolute_error(pred, y_test))

# r2_score
print(r2_score(pred, y_test))
```

```python
# 잔차
residual = pred - y_test

print(residual)
```

```python
'''
잔차가 정규분포를 따르는지 시각적으로 확인하는 Q-Q plot.
점들이 직선 근처에 모이는 형태 = 정규성 만족
점들이 곡선 형태로 퍼지는 형태 = 정규성 위배
'''
import scipy.stats as stats

plt.figure(figsize=(5,3))
stats.probplot(residual, dist='norm', plot=plt)

plt.tight_layout()  # 그래프 요소 간격 자동 정리
plt.show()
```
